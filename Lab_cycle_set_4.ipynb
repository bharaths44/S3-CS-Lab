1. Draw a Dot Chart for 3 variables for the data.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(20)
y = np.random.rand(20)
z = x*y

fig = plt.figure(figsize=(6, 6))
plt.scatter(x, y,
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 200,
           c=z)
plt.show()
```

2. Draw a Pie Chart and Radial Diagram for the data.

```python
import matplotlib.pyplot as plt

data = [25, 25, 25, 25]
labels = ['A', 'B', 'C', 'D']

# Pie Chart
plt.pie(data, labels=labels)
plt.show()

# Radial Diagram
ax = plt.subplot(111, polar=True)
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
ax.set_rlabel_position(0)
ax.set_yticklabels([])
ax.set_xticklabels(labels)
ax.plot(data)
ax.fill(data, 'b', alpha=0.1)
plt.show()
```

7. Draw a Scatter Plot for Outliers Highlighted for the data.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
x = np.random.normal(50, 20, 100)
y = np.random.normal(50, 20, 100)
x[-10:] = x[-10:] + 200
y[-10:] = y[-10:] + 200
lower_bound_x = x.mean() - 3 * x.std()
upper_bound_x = x.mean() + 3 * x.std()
lower_bound_y = y.mean() - 3 * y.std()
upper_bound_y = y.mean() + 3 * y.std()
outliers_x = x[(x < lower_bound_x) | (x > upper_bound_x)]
outliers_y = y[(y < lower_bound_y) | (y > upper_bound_y)]
plt.scatter(x, y, color='blue', alpha=0.5)
plt.scatter(outliers_x, outliers_y, color='red')
plt.show()
```

***8. Draw a Scatter Plot for Areas Highlighted for the data.***

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(100, 2)

# Define the areas to highlight
area1 = np.where((data[:, 0] >= 0.3) & (data[:, 0] <= 0.6) & (data[:, 1] >= 0.3) & (data[:, 1] <= 0.6))
area2 = np.where((data[:, 0] >= 0.7) & (data[:, 0] <= 1) & (data[:, 1] >= 0.7) & (data[:, 1] <= 1))

# Create the scatter plot
plt.scatter(data[:, 0], data[:, 1], color='blue')
plt.scatter(data[area1[0], 0], data[area1[0], 1], color='green', marker='x')
plt.scatter(data[area2[0], 0], data[area2[0], 1], color='red', marker='x')

# Add labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```
