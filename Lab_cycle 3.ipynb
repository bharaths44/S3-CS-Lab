**AGGREGATION AND GROUPING**

**Qn.1 Load 'planets' dataset and apply simple aggregation functions to the dataset that grouped by the feature 'method' and iterate over the group**

* count() Total number of items
* first(), last() First and last item
* mean(), median() Mean and median
* min(), max() Minimum and maximum
* std(), var() Standard deviation and variance
* mad() Mean absolute deviation
* prod() Product of all items
* sum() Sum of all items




```python
import seaborn as sns
planets = sns.load_dataset('planets')
planets.shape
```

```python
planets.head()
```

```python
planets.groupby('method')['orbital_period'].median()
```

```python
for (method, group) in planets.groupby('method'):
  print("{0:30s} shape={1}".format(method, group.shape))
```

****Qn.2 Apply groupby operation on the feature 'method' with aggregation, filtering, transform and apply to the planets dataset.**

* 1. Find the minimum of all attributes
* 2. Filter the values whose year value count is greater than 5
* 3. Transform the values to a new value that is subtracted with its mean
* 4. Apply the same filter function to the dataset using 'apply'

```python
planets.groupby('method').aggregate('min')
```

```python
def filter_func(x):
  return x['year'].count() > 5
```

```python
print(planets.groupby('method').count());
print(planets.groupby('method').filter(filter_func)['method'].unique())

```

```python
planets.groupby('method').transform(lambda x: x - x.mean())
```

```python
print(planets.groupby('method').apply(filter_func))
```
